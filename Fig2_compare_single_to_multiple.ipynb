```python
def linear_stretch(band, lower=2, upper=98):
    lower_val = np.percentile(band, lower)
    upper_val = np.percentile(band, upper)
    return np.clip((band - lower_val) / (upper_val - lower_val), 0, 1)

savepic=os.path.join(fig_path,'Fig2_comparisons_between_single_and_yearly_means.png')

LakeName1='WuhaiLake'
file_list1=['5years_mean','2021-12-17','2021-12-02','2021-05-04']
LakeName2='SanmenRser'
file_list2=['5years_mean','2020-06-14','2020-06-29','2020-08-28']

plt.figure(figsize=(8,4),dpi=300)#4*4
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=0.05,hspace=0.05)
for i in range(len(file_list)):
    #Lake1
    image = rasterio.open(os.path.join(image_dir, f'{LakeName1}_{file_list1[i]}.tif'))
    read_image = image.read()
    transform = image.transform
    ax1=plt.subplot(4,4, i+1)
    rgb=read_image[[2, 1,0], :, :]
    rgb_stretched = np.array([linear_stretch(band) for band in rgb])
    show(rgb_stretched,transform=transform,ax=ax1)
    ax1.set_xticks(np.arange(transform[2], transform[2] + image.shape[1]*0.00009, 0.1))
    ax1.set_yticks(np.arange(transform[5]-0.01, transform[5] + image.shape[0]*-0.00009, -0.1))

    ax1.tick_params(axis='y',labelright=True,labelleft=False)
    plt.tick_params(bottom=True, top=True, left=True, right=True )
    plt.tick_params(labelbottom=False,labeltop=True,labelleft=True, labelright=False)
    ax1.xaxis.set_major_formatter('{x:.2f}°E')  # 格式化标签
    ax1.yaxis.set_major_formatter('{x:.2f}°N')
    
    ax2=plt.subplot(4,4, i+5)
    img = show(read_image[-2],transform=transform, ax=ax2,vmin=-10000,vmax=10000,cmap='RdBu')
    if i==3:
         plt.colorbar(img.get_images()[0],ax=ax2)
            
    ax3=plt.subplot(4,4, i+9)
    image = rasterio.open(os.path.join(image_dir, f'{LakeName2}_{file_list2[i]}.tif'))
    read_image = image.read()
    transform = image.transform
    rgb=read_image[[2, 1, 0], :, :]
    rgb_stretched = np.array([linear_stretch(band) for band in rgb])
    show(rgb_stretched,transform=transform,ax=ax3)
    
    ax3.set_xticks(np.arange(transform[2], transform[2] + image.shape[1]*0.00009, 0.1))
    ax3.set_yticks(np.arange(transform[5]-0.01, transform[5] + image.shape[0]*-0.00009, -0.1))

    ax3.tick_params(axis='y',labelright=True,labelleft=False)
    plt.tick_params(bottom=True, top=True, left=True, right=True )
    plt.tick_params(labelbottom=False,labeltop=True,labelleft=True, labelright=False)
    ax3.xaxis.set_major_formatter('{x:.2f}°E')  # 格式化标签
    ax3.yaxis.set_major_formatter('{x:.2f}°N')
    
    ax4=plt.subplot(4,4, i+13)
    img = show(read_image[-2],transform=transform, ax=ax4,vmin=-10000,vmax=10000,cmap='RdBu')
    if i==3:
         plt.colorbar(img.get_images()[0],ax=ax4)
        
plt.savefig(savepic,dpi=400,bbox_inches='tight')#,dpi=400
plt.show()
```
